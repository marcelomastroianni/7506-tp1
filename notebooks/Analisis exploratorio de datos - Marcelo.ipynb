## Analisis Exploratorio Usos de los Edificios


```python
df_joined = df_train_values.merge(df_train_labels, left_on='building_id', right_on='building_id', how='inner')
```

```python
df_joined.head(10)
```

Me quedo con las columnas que indican los distintos usos que tienen los edificios

```python
df_building_uses =df_joined[["has_secondary_use_agriculture","has_secondary_use_hotel","has_secondary_use_rental","has_secondary_use_institution","has_secondary_use_school","has_secondary_use_industry","has_secondary_use_health_post","has_secondary_use_gov_office","has_secondary_use_use_police","has_secondary_use_other","damage_grade"]]
df_building_uses
```

```python
df_building_uses_grouped = df_building_uses.groupby(["damage_grade"]).sum()
df_building_uses_grouped
```

```python
df_building_uses_grouped2 = df_building_uses_grouped.stack().reset_index()
df_building_uses_grouped2
```

```python
table = pd.pivot_table(df_building_uses_grouped2, values=0, index=['level_1'],columns=['damage_grade'], aggfunc=np.sum)
table
```

```python
ax = table.plot.barh(rot=0)

```

Del analisis del grafico podemos ver que de las estructuras que "tenian un uso alternativo" las mayores estructuras dañadas fueron las de uso para agricultura, seguidas de las de uso como hoteles.
De las estructuras de agricultura las mayoria recibieron un daño de grado 2 seguidas de un daño de grado 3, mientras que las estructuras usadas como hoteles tambien tuvieron mas daños con grado 2 seguidas de daños de grado 1.
